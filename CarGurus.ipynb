Time to find the class that I need to scrape

```python
#Getting the class names that are available
tags = {tag.name for tag in soup.find_all()}

class_list = set()
for tag in tags:
  
    # find all element of tag
    for i in soup.find_all( tag ):
  
        # if tag has attribute of class
        if i.has_attr( "class" ):
  
            if len( i['class'] ) != 0:
                class_list.add(" ".join( i['class']))
  
print( class_list )
```

After a long search, I found that 'TitleWrap' had what I was looking for

```python
whole_listings = soup.find_all('h4')
whole_listings
```

Now to extract the text portion from the whole_listings and pass it to a list: text_list

```python
text_list = []
for div in soup.findAll('div',{'class':'titleWrap'}):
    text_list.append(div.text.strip())
```

# Creating the dataset

Converting the text_list to a DataFrame: data

```python
data = pd.DataFrame(text_list)
```

## Creating Empty Columns

```python
#Creating empty columns to be filled
cols = ["Year","Manufacturer", "Model", "Price", "Mileage", "Certified Pre-owned?", "Color"]
for col in cols:
    data[col] = ''
```

```python
#Verification
data
```

## Cleaning the rows

```python
for index,row in data.iterrows():
    data[0][index] = re.sub('\n','',data[0][index]) #Getting rid of the '\n'
    data[0][index] = re.sub('\xa0','',data[0][index]) #Getting rid of the '\xa0'
    data[0][index] = re.sub(' +', ' ', data[0][index]) #Getting rid of the extra spaces
```

```python
#Checking the output
data.head()
```

Looks good! Now onto the next

## Splitting the columns

```python
for index, row in data.iterrows(): #Iterating over the rows
    
    # Splitting the words in the text to a list of words
    word_list = data[0][index].split()
    #print(word_list)
    
    #Taking the first word as the Year, in numeric form
    data['Year'][index] = int(word_list[0])
    
    #Taking the second word as the Manufacturer Name
    data['Manufacturer'][index] = word_list[1]
    
    ######NOTE : Look into two word Manufacturer names, like Mercedes-Benz , Land Rover, Range Rover and Alfa Romeo
    
    
    #Model name is the text that precedes "Description", so adding all the words that precede it to the 'M-odel'
    model_name = ''
    wrd_ctr = 2 
    while(word_list[wrd_ctr] != "Description:"):
        model_name = model_name + ' ' +  word_list[wrd_ctr]
        #print(model_name)
        data['Model'][index] = model_name
        wrd_ctr += 1
    
    #Price: Searching for the element which has a '$' sign
    data["Price"][index] = [i for i in word_list if i.startswith('$')][0] #Finding the word that starts with a $ sign. The [0] at the end is to unpack the list that contains only one element.
    price_index = word_list.index([i for i in word_list if i.startswith('$')][0])  #Saving the index of the price for use in mileage
    data["Price"][index] = re.sub(',','',data["Price"][index]) #Removing the ','
    data["Price"][index] = re.sub('\$','',data["Price"][index]) #Removing the '$' sign
    data["Price"][index] = int(data["Price"][index]) #Converting to integer
    
    #Mileage: Searching for the word that lies in two positions after "Price"
    data["Mileage"][index] = word_list[price_index+2]
    data["Mileage"][index] = re.sub(',','',data["Mileage"][index]) #Removing the ','
    data["Mileage"][index] = int(data["Mileage"][index]) #Converting to integer
    
    #Certified Pre-Owned?
    word_index = word_list.index([i for i in word_list if i.startswith('Pre-Owned')][0])
    #print(word_list[word_index])
    data["Certified Pre-owned?"][index] = word_list[word_index].split(':')[1] # Splitting the phrase "Pre-Owned: Yes/No" with ':' to get the response
    
    #Color
    try:
        color_index = word_list.index([i for i in word_list if i.endswith('Color:')][0])
        #print(word_list[color_index])
        color =''
        for i in range(color_index+1, len(word_list)):
            color = color + ' ' + word_list[i]
            data["Color"][index] = color
    except:
        data["Color"][index] = "NA"
    
    
```
